<h1>Python Print Tricks 🔥 Concatenation vs Comma vs .format()</h1>

<h3>🔹 1. Concatenation (+)</h3>
Works only with same type (string + string)

```python
a="python"
b=5
print("hi " + str(b))
```

<h3>🔹 2. Comma (,)</h3>
<li>Automatically adds space</li>
<li>No type issue 👍</li>

```python
a="python"
b=5
print("hi" , b)
```

<h3>3. .format()</h3>
<p>It is string method</p>
<p>insert variables or expressions into a string through placeholders defined by curly braces {}</p>

```python
a="python"
b=5
print("Hi {}".format(b))
```

<h4> Index based format</h4>

```python
a="python"
b=5
print("Hi {1} i am {0}".format(a,b))
```

<h4>Variable name based </h4>

```python
print("Hi i am {a} my age is {age}".format(a="banu",age=37))
```

<h4>How to evaluate expression</h4>

```python
a=5
b=10
print("Multiplication of a and b is {}".format(a*b))
```

<h4>how to evaluate function</h4>

```python
a="Python"
print("Length of python {}".format(len(a)))
```
