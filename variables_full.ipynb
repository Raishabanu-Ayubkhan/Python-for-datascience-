# what is variable?
<p style="color: blue; font-size:20px">“In Python, variables are symbolic names that refer to objects stored in memory.”</p>

```python
# assign 5 and check type
x=5
print(x)
print(type(x))
```

1. 5 is an object ✅
2. x is a reference (name) to that object ✅

**Variable Naming Rules**
  

      
   
   

```python
#1. Must start with letter or underscore
name="Python"
_name="RB coding school"
print(name)
print(_name)
```

```python
num2=67
print(num2)
```

```python
#2. Cannot use keywords
if="Computer"
print(if)
```

```python
#3. No special characters
num@1=5
num-3=6
print(num@1)
print(num-3)
```

```python
#4. Case sensitive
name=78
Name=64
print(name)
print(Name )
```

```python
#5. Can contain letters, numbers, underscore
mark_1=75
print(mark_1)
```

# Multiple Variables 

  Assigning many variables in one line.

```python
a,b,c=1,2,3
print(a,b,c)
```

# Multi-value Assignment (same value)


```python
a=b=c=10
print(a,b,c)
```

# Swapping variable

```python
a=10
b=20
print(a,b)
a,b=b,a
print(a,b)
```

# unpacking


```python
a,b,c="hai"
print(a,b,c)
```

# starred variable(*a)----> collecting elements to make list
# it will not work all alone


```python
a,*b="hai"
print(a,b)
```

```python
a,b,*c="Python"
print(a,b,c)
```
