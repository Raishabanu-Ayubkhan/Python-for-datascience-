## 📌 What are Data Types?
- A data type defines the type of value a variable holds.  
- In python data types are classes and not declared by user
- Interpreter determine at run time based on assigned value 


<h2>Python Data Types</h2>

<img src="data type.png" alt="Python Data Types Diagram" width="600">


| Name           | Type    | Description                                           |
| -------------- | ------- | ----------------------------------------------------- |
| Integers       | int     | Whole numbers: `3`, `300`, `-200`                     |
| Floating point | float   | Numbers with decimal point: `2.3`, `4.6`, `100.0`     |
| Complex        | complex | Numbers with real and imaginary parts: `2+3j`, `1-4j` |
| Strings        | str     | Ordered sequence of characters: `"hello"`, `"Sammy"`  |
| Lists          | list    | Ordered, mutable collection: `[10, "hello", 200.3]`   |
| Dictionaries   | dict    | Key-value pairs: `{"name": "Frankie"}`                |
| Tuples         | tuple   | Ordered, immutable collection: `(10, "hello", 200.3)` |
| Sets           | set     | Unordered collection of unique values: `{"a", "b"}`   |
| Booleans       | bool    | Logical values: `True`, `False`                       |


## 🔢 Numeric Types

# Integer (int)
**Whole numbers: 3, 300, -200**

```python
x=5
y=-10
z=0
print(x,y,z)
print(type(x),type(y),type(z))
```

# Floating point (float)
**Numbers with decimal point: 2.3, 4.6, 100.0**

```python
a=2.3
b=-4.5
c=0.0
print(a,b,c)
print(type(a),type(b),type(c))
```

# Complex Number (complex)
**Numbers with real and imaginary parts: 2+3j, 1-4j**


```python
comp=3+7j
print(comp)
print(comp.real)
print(comp.imag)
```

## 📚 Sequence Types

## 🔤 String (str)

## 🔤 List (list)
**Ordered, mutable collection: [10, "hello", 200.3]**

```python
list_1=[5,"banu",4.5]
list_1[0]=10
type(list_1)
```

## 🔤 Tuple (tuple)
**Ordered, immutable collection: (10, "hello", 200.3)**

## 🔘 Boolean (bool)
**Logical values: True, False**

```python
is_true=False
is_true
type(is_true)
```

## 🧩 Set (set)
**Unordered collection of unique values: {"a", "b"}**

```python
set_1={1,2,5,4,1,3,2}
set_1
```

```python
type(set_1)
```

## 🗂️ Dictionary (dict)
**Key-value pairs: {"name": "Frankie"}**

```python
dict_1={"name":"banu","age":37,"marks":80.7}
dict_1
print(type(dict_1))
dict_1.keys()
dict_1.values()
dict_1["name"]
```
