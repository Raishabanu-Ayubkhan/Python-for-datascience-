<h2>🔹 What are Escape Sequences in Python? </h2>

<p>Escape sequences are special characters inside strings that start with a backslash \.

They are used when you want to:
<ul>
    <li>Print something special (like new line, tab)</li>
    <li>Include quotes inside strings</li>
    <li>Format output nicely</li>
</ul>
</p>

<h4>1. New Line → \n </h4>

Moves text to the next line.

```python
print("hello\nworld")
```

<h4> 2. Tab Space → \t </h4>

Adds a tab (space).

```python
print("A\tB")
print("AAA\tB")
```

3. Backslash → \\   

Prints a single backslash.

```python
print("c:\\Admin\\destop\\comm.py")
```

<h4> 4. Single Quote → \' </h4>

```python
print('It\'s python')
```

<h4>5. Double Quote → \"</h4>


```python
print("Hi i am \"banu\"")
```

<h4>6. Carriage Return → \r</h4>

Moves cursor to beginning (overwrites text).

```python
print("Hello\rhai")
```

<h4>7. Backspace → \b</h4>

Removes previous character.

```python
print("Hai\b")
```

<h2>Raw String </h2>

<p>If you don’t want escape sequences to work, use raw string </p>
Syntax : (r"")

```python
print(r"c:\Admin\destop\comm.py   ")
print(r"hello\'\''\r")
```

```python
print("Name:\tbanu\nage:\t37")
```
