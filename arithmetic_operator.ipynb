<h1>Operators</h1>
<h4>In Python, an operator is a special symbol or keyword used to perform operations on values and variables.</h4>
<h4>The values that an operator acts upon are called operands</h4>
<p>For example, in the expression 10 + 5, the + is the operator and 10 and 5 are the operands.</p>

<h3> Arithmetic operators +,-,*,/,%,//,** </h3>

<h4> +(plus) -> Addition </h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1+num_2)
```

<h4> -(minus) -> Subtraction</h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1-num_2)
```

<h4> *(asterisk) -> Multiplication </h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1*num_2)
```

<h4> /(slash) -> Division returns float quotient (output= 5.0) </h4>
<img src="division.jpeg">

```python
num_1=int(input())
num_2=int(input())
print(num_1/num_2) 
```

<h4> %(modulo) -> Division returns remainder </h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1%num_2)
```

<h4> //(double slah) -> floor Division returns integer quotient </h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1//num_2)
```

<h4> **(Double asterisk) -> exponentiation base**exponent / power operator </h4>

```python
num_1=int(input())
num_2=int(input())
print(num_1**num_2)
```
