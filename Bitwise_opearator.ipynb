<h1>Bitwise Opeartor</h1>
<h3>Bitwise operators in Python perform operations directly on the 
    binary representations (0s and 1s) of integers</h3>
<h3>👉 “Bitwise operators only work with integers, not floats.”</h3>   

<h3>Python provides six primary bitwise operators</h3>

<h3> Bitwise AND  symbol(&)- ampersand </h3>
<h2> AND Truth table</h2>
<img src = "bit_and.png">

```python
# Example of Bitwise AND
print(7&5)
```

<h3>Bitwise OR symbol(|) pipe or vertical bar </h3>
<h3> OR Truth table </h3>
<img src="bit_or.png">

```python
print(7|5)
```

```python
Bitwise OR Execution
7  = 111
5  = 101
------------
|  = 111 
------------
Convert binary to decimal integer 
      1     1     1
      |     |     |
      |     |     |____ 1 × 2^0 = 1
      |     |__________ 1 × 2^1 = 2
      |________________ 1 × 2^2 = 4

Total = 4 + 2 + 1 = 7
```

```python
# How to convert integer to binary 
print(bin(5)[2:])
```

```python
# How to convert binary to integer
int("101",2)
```

<h3> Bitwise XOR symbol (^) caret </h3>
<h4>XOR Truth table </h4>
<img src ="bit_xor.png">

<h4>👉 XOR stands for “Exclusive OR”</h4>
<h4>🧠 What does “exclusive” mean?

👉 It means:

Only one should be true — not both</h4>

```python
print(7^5)
```

```python
Bitwise XOR Execution
7  = 111
5  = 101
------------
^  = 010 
------------
Convert binary to decimal integer 
     1     0
     |     |
     |     |____ 0 × 2^0 = 0
     |__________ 1 × 2^1 = 2


Total = 0+ 2  = 2
```

<h3>Bitwise NOT symbol(~) tilde </h3>
<h4> NOT Truth table </h4>
<img src= "bit_not.png">

```python
print(~5)
print(bin(-6 & 0xFF)[2:])
```

```python
binary of 5 in 8 bit   101
                         0000 0101
taking not             ~ 1111 1010

left most bit = 1 (-negative)
left most bit = 0 (+positive)

o/p = 1111 1010  left most is 1 so output is negative number

What number is that?
 2 steps
                            1111 1010
step 1: 1's complement      0000 0101
step 2: 2's complement             +1
                            ----------
                            0000 0110
                            ----------

decimal(0000 0110) --> 6
already we know output is negative so -6
```

<h3>Left Shift operator << </h3>
<h2>👉 When you do left shift, you add zeros at the end (right side)</h2>

```python
print(5<<2)
```
