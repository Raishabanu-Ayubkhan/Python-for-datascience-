<h1>f-String </h1>
<p>👉 “f-string means formatted string.</p>
<p>It allows you to do following inside curly braces </p>
<li>Values</li>
<li>Variables</li>
<li>expression</li>
<li>Functions</li>

<h3>Syntax</h3>
f" {} "
<p>important note : f should be before quotes</p>

<h3>Values</h3>

```python
print(f"My mark is {85}")
```

<h3>Variables</h3>

```python
pi=3.1415
print(f"The value of pi is {pi}")
```

<h3>Expression</h3>

```python
print(f"The value of pi is {22/7}")
```

<h3>Function</h3>

```python
name="python"
print(f"Name is {name.upper()}")
```

```python
def add(a,b):
    return(a+b)
print(f"The addition of 4 and 5 is {add(4,5)}")
```

<h1> Format specifier </h1>
<h3>Syntax</h3>
f"{value:format_spec}"

<p>👉 : starts formatting</p>
<p>👉 format_spec = how you want output</p>

<h3>:.(f,%,e)</h3>

<h4>Fixed floating point :.f </h4>

```python
print(f"VAlue of pi is {22/7}")
print(f"VAlue of pi is {22/7:.3f}")
```

<h4>Percentage  :.%</h4>

```python
avg=0.856
print(f"percentage average is {avg:.2%}")
```

<h4>Scientific Notation :.e</h4>

```python
avg = 9.3
print(f"The average {avg:.2e}")
```

<h4>width          :width</h4>

```python
name="python"
print(f"{name:15}")
```

<h4>Width + decimal  :width.decimal </h4>

```python
avg=3.1456783
print(f"{avg:15.2f}")
```

<h4>Alignment</h4>

```python
avg=3.1456783
print(f"{avg:^15.2f}")
```

<h4>Padding</h4>

```python
avg="banu"
print(f"{avg:*^15}")
```
