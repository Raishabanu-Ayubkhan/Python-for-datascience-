# 🧠 Python Input Function

✨ Learn how to take input from users in Python  
🎯 This is the first step to make programs interactive!

## 📌 What is input()?

👉 `input()` is used to take data from the user.

💡 It always returns the data as **string (str)** by default.

## ⚙️ Syntax

```python
input(prompt)

🔹 prompt → Message shown to the user
🔹 Returns → Always string

````python
#Basic Example
name=input("what is your name?")
print(name)
````

## ⚠️ Important Note

Even if you enter a number, Python treats it as **string**!

````python
num_1=input("Enter the number")
type(num_1)
#print(num_1+10)
````

## 🔄 Type Conversion

To use numbers, convert input using:

- `int()` → whole numbers  
- `float()` → decimal numbers

````python
a=int(input("Enter the value for a"))
print(type(a))
print(a+10)
````

## 👥 Multiple Inputs

````python
a=int(input("Enter a"))
b=float(input("Enter b"))
print(a+b)
````

## ⚡ Multiple Values in One Line

````python
a,b=input("Enter the value for a and b").split()
print(type(a))
print(b)
print(int(a) + int(b) )
````

## ⚡ eval

````python
total=eval(input("Enter any expression"))
print(total)
````
