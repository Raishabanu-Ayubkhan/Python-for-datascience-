# print() is a built-in function used to display output on the screen.

# Basic print() Syntax
**print(object(s), sep=' ', end='\n', file=sys.stdout, flush=False)**

```python
Explanation of all 5 parameters:
object(s)
ЁЯСЙ What you want to print (text, number, variable, etc.)
sep (separator)
ЁЯСЙ Used when printing multiple values
ЁЯСЙ Default = space ' '
end
ЁЯСЙ What to print at the end
ЁЯСЙ Default = newline \n
file
ЁЯСЙ Where to print output (screen or file)
flush
ЁЯСЙ Forces output immediately (advanced use)
```

```python
# Basic Example only objects 
print("hello")
print(2)
a=7.5
print(a)
```

```python
#Using sep Parameter тнР  default sep =" "
print("hello","world",sep='тнР')
date=21
month="08"
year=1987
print(date,month,year,sep='/')
```
