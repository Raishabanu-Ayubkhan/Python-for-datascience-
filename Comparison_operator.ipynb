<h1>Comparison Operator or Relational operator</h1>
<h3> It compares 2 values and return boolean(True or False) as output</h3>

<h3>Comparison with numerical Value</h3>

```python
#        ==	       Equal to	
num_1=int(input())
num_2=int(input())
print(num_1==num_2)
```

```python
# !=	       Not equal to	
print(2!=2)
```

```python
# >	       Greater than	 
print(5>2)
print(2>5)
```

```python
# <	       Less than
print(2<5)
print(5<2)
```

```python
# >=	       Greater than or equal to
print(1>=2)
```

```python
# <=	       Less than or equal to
print(1<=2)
print(2<=2)
print(2>=5)
```

<h3>Comparison with string value </h3>
<p>String comparison is based on ASCII value </p>

<h4>ASCII value for '0'=48,'1'=49 to '9'=57</h4>
<h4>ASCII value for 'A'=65 , 'a'=97</h4>

```python
#        ==	       Equal to	
str_1=input()
str_2=input()
print(str_1==str_2) 
```

```python
# !=	       Not equal to	
print("apple"!="Apple")
```

```python
# >	       Greater than	 
print("car">"cat")
```

```python
# <	       Less than
print("Zoo"<"apple")
```

```python
# >=	       Greater than or equal to
print("abc">="ABC")
```

```python
# <=	       Less than or equal to
print('10'<='2')
```
